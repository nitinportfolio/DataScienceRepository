# Series
- It is similar to arrasy except we can name the index 

```python
mydata = [10,20,30,40,50]
```

```python
mylabels = ['L1','L2','L3','L4','L5']
```

```python
myser = pd.Series(mydata)
```

```python
myser
```

```python
myser = pd.Series(mydata, index=mylabels)
```

```python
myser
```

```python
mydata = [100,200,300,400]
myindexes = ['mobilephone', 'headphone', 'earphone', 'microphone']
```

```python
myser = pd.Series(mydata, index= myindexes)
```

```python
myser
```

```python
myser[['mobilephone','earphone']]
```

```python
myser>200
```

```python
myser[myser>200]
```

```python
type(myser)
```
