# Logistic Regression Classifier Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train.shape
```

```python
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
y_predict = model_LR.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

The recall means "how many of this class you find over the whole number of element of this class"<br>

The precision will be "how many are correctly classified among that class"<br>

The f1-score is the harmonic mean between precision & recall<br>

The support is the number of occurence of the given class in your dataset <br>(so you have 37.5K of class 0 and 37.5K of class 1, which is a really well balanced dataset.<br>

The thing is, precision and recall is highly used for imbalanced dataset <br>because in an highly imbalanced dataset, a 99% accuracy can be meaningless.<br>

I would say that you don't really need to look at these metrics for this<br> problem , unless a given class should absolutely be correctly determined.<br>

To answer your other question, you cannot compare the precision and the<br> recall over two classes. This only means you're classifier is better to find class 0 over class 1.<br>

Precision and recall of sklearn.metrics.precision_score or recall_score <br>should not be different. But as long as the code is not provided, this is impossible to determine the root cause of this.

If the drove of fishes is huge, while the net is pretty small -> We will see fishes in very positions in the net, means Precision is high. But we only get a minority of the drove, means Recall is low.

Meanwhile, there is just a tiny drove of fishes, but we got a huge net -> We will see that just a tiny part of the net have fishes, means Precision is low. But fortunately, we catch every fish in the drove, means Recall is high.
