# Distplot or Distribution Plot
distplot - is used to see how data is distributed. Output will be in the form of histogram with the KDE (kernel density estimation ) line

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# JointPlot


```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips,kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips,kind = 'reg')
```

# Pairplot 
- It shows relationship of entire dataset for all numerical columns

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex',diag_kind = 'hist',palette = 'deep')
```

```python
tips.head()
```

```python
sns.pairplot(tips, hue = 'smoker',diag_kind = 'hist',palette = 'deep')
```
