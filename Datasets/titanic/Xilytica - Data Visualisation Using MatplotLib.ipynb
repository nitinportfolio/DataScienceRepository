# Customising the plot for colors, line types etc

```python
fig = plt.figure(figsize=(8,4))
axes = fig.add_axes([0,0,1,1])
# Colors - orange, green, Yellow etc
# We can also add hex code
# Line width or lw- default is 1
# Alpha - gives the transparent the line is
# linestyle or ls - --, -. etc
# Markers - points on the line - *, +, o, 1, 2 etc
# markersize - to increase the size of marker
# axes.plot(x,y**2, label = ' Curve', color = 'orange')
axes.plot(x,y**2, label = ' Curve', color = '#b34180', linewidth = 2, alpha = .8, linestyle = '--',
          marker = 'o', markersize = 10, markerfacecolor = 'yellow', markeredgewidth = 2, markeredgecolor = 'blue')

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('My New Plot')

axes.legend(loc = 0)
```

```python
fig = plt.figure(figsize=(8,4))
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y**2, label = ' Curve', color = '#ec4646')

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('My New Plot')

axes.legend(loc = 0)
```

```python
# Want to show graph between 0 and 2 - means zooming the plot
# give list with lower & upper bound
# xlim & ylime
```

```python
fig = plt.figure(figsize=(8,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y**2, label = ' Curve', color = '#ec4646')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('My New Plot')

axes.legend(loc = 0)
axes.set_xlim([0,2])
axes.set_ylim([0,10])

axes.spines['bottom'].set_color('#dddddd')
axes.spines['top'].set_color('#dddddd') 
axes.spines['right'].set_color('red')
axes.spines['left'].set_color('red')

axes.tick_params(axis='x', colors='red')
axes.tick_params(axis='y', colors='red')

axes.yaxis.label.set_color('red')
axes.xaxis.label.set_color('red')
```
