```python
import numpy as np
import plotly.offline as pyo
import plotly.graph_objects as go
import pandas as pd
```

```python
# Scatter Plots 
# Allows comparison of two variables for a set of data. One variable on X axis & other on Y axis
# Used to see the corretaltion in variables
```

```python
np.random.seed(101)
```

```python
random_x = np.random.randint(1,101,100)
random_y = np.random.randint(1,101,100)
```

```python
data = [go.Scatter(x = random_x, y= random_y, mode = 'markers')]
```

```python
pyo.plot(data, filename='scatter.html')
```

```python
# Adding titles & axes names 
data = [go.Scatter(x = random_x, y= random_y, mode = 'markers')]
layout = go.Layout(title = 'Scatter Plot',
                  xaxis = dict(title = 'My X Axis'),
                  yaxis = dict(title='My Y Axis'),
                  hovermode='closest')
fig = go.Figure(data = data, layout = layout)
pyo.plot(fig, filename='scatter.html')
```

```python
# Adding titles & axes names 
data = [go.Scatter(x = random_x, y= random_y,
                   mode = 'markers',
                   marker = dict(size=14,
                                color = '#3f3697',
                                symbol = 'pentagon',
                                line={'width':2}))]
layout = go.Layout(title = 'Scatter Plot',
                  xaxis = dict(title = 'My X Axis'),
                  yaxis = dict(title='My Y Axis'),
                  hovermode='closest')
fig = go.Figure(data = data, layout = layout)
pyo.plot(fig, filename='scatter.html')
```

```python
# Line Chart
# Dispalys the series of data points connected by line segments
# Often used to visualise trend overt he time line
```

```python
import numpy as np
import plotly.offline as pyo
import plotly.graph_objects as go


np.random.seed(101)
x_values = np.linspace(0,1,100)
y_values = np.random.randn(100)

trace0 = go.Scatter(x = x_values, y = y_values+5,
                    mode = 'markers',name = 'markers')

trace1 = go.Scatter(x = x_values, y = y_values,
                    mode = 'lines',name = 'mylines')

trace2 = go.Scatter(x = x_values, y = y_values-5,
                    mode = 'lines + markers',name = 'mylines & mymarkers')
data = [trace0,trace1,trace2]

layout = go.Layout(title = 'Line Charts')
fig = go.Figure(data = data, layout=layout)
pyo.plot(fig)

```

```python
# Bubble Chart
# Similar to scatter plot, except that we can get information of third variable through the size of the marker
# We can add more variable  informations by coloring the points based on categories
```

```python
data = pd.read_csv(r'C:\Users\Administrator\Class Notebooks\data\cars.csv')
```

```python
data.head()
```

```python
# multiplied cyl by 5 to get big size bubbles
datagr = [go.Scatter(x = data['hp'],
                     y = data['mpg'],
                     text =data['Unnamed: 0'],
                     mode = 'markers',
                     marker=dict(size = 5*data['cyl']))]
layout = go.Layout(title = 'Bubble Chart')
#datagr = [trace1,trace2,trace3]
fig = go.Figure(data = datagr,layout = layout)
pyo.plot(fig)
```

```python
# multiply wt by 10 to get bigger size bubbles
datagr = [go.Scatter(x = data['hp'],
                     y = data['mpg'],
                     text =data['Unnamed: 0'],
                     mode = 'markers',
                     marker=dict(size = 10*data['wt'], 
                     color = data['cyl'],
                                showscale = True))]
layout = go.Layout(title = 'Bubble Chart', hovermode = 'closest')
#datagr = [trace1,trace2,trace3]
fig = go.Figure(data = datagr,layout = layout)
pyo.plot(fig)
```

```python
# Histogram - Shows distribution of a continuous variable
# Bins are intervals
```

```python
datagr = [go.Histogram(x = data['mpg'])]
layout = go.Layout(title = 'My Histogram')
fig = go.Figure(data = datagr, layout = layout)
pyo.plot(fig)
```

```python
datagr = [go.Histogram(x = data['mpg'], xbins = dict(start = 0, end = 30, size = 2))]
layout = go.Layout(title = 'My Histogram')
fig = go.Figure(data = datagr, layout = layout)
pyo.plot(fig)
```

```python
# Heatmaps - We can visualize 3 features 
# Categorical or Continuous features along x & y axis to make up a grid &
# 3rd continuous feature displayed through color
```

```python
data = pd.read_csv('C:/Users/Administrator/visualizations/data/santabarbara.csv')
```

```python
data.head()
```

```python
# z = needs to be python list so we doing the transformation
datagr = [go.Heatmap(x = data['DAY'], y = data['LST_TIME'], 
                     z = data['T_HR_AVG'].values.tolist(),
                    colorscale = 'Jet')]
layout = go.Layout(title = 'Santabarbara Temprature HeatMap')
fig = go.Figure(datagr, layout)
pyo.plot(fig)
```

```python
data1 = pd.read_csv('C:/Users/Administrator/visualizations/data/SantaBarbara.csv')
data2 = pd.read_csv('C:/Users/Administrator/visualizations/data/Yuma.csv')
data3 = pd.read_csv('C:/Users/Administrator/visualizations/data/Sitka.csv')
```

```python
data2.head()
```

```python
from plotly import tools
```

```python


trace1 = go.Heatmap(x = data1['DAY'], y = data1['LST_TIME'], 
                     z = data1['T_HR_AVG'].values.tolist(),
                    colorscale = 'Jet', zmin = 5, zmax = 40)



trace2 = go.Heatmap(x = data2['DAY'], y = data2['LST_TIME'], 
                     z = data2['T_HR_AVG'].values.tolist(),
                    colorscale = 'Jet', zmin = 5, zmax = 40)

trace3 = go.Heatmap(x = data3['DAY'], y = data3['LST_TIME'], 
                     z = data3['T_HR_AVG'].values.tolist(),
                    colorscale = 'Jet', zmin = 5, zmax = 40)

fig = tools.make_subplots(rows=1, cols = 3, subplot_titles = ['Santabarbara','Yuma','Sitka'], shared_yaxes = True)

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)
fig.append_trace(trace3,1,3)

fig['layout'].update(title ='Temps For 3 Cities')
pyo.plot(fig)
```
