# Group By 
- GroupBy - Function is used to split the data into groups based on some criteria & apply some aggregate function on them. Pandas objects can be split on any of their axes.

```python
new_data = {
    'Company':['SamPhone','SamPhone','MIPhone','MIPhone','iPhone','iPhone'],
    'Person':['Jack','Tim','Jenifer','Amy','Sarah','Bob'],
    'Qty':[10,16,12,30,10,12],
    'Sales':[1000,2000,3000,4000,5000,6000]
    
}
```

```python
df = pd.DataFrame(new_data)
```

```python
df
```

```python
df.groupby('Company').mean()
```

```python
grp = df.groupby('Company')
```

```python
grp.mean()
```

```python
df
```
