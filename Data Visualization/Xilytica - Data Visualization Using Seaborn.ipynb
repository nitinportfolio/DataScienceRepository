# Distplot or Distribution Plot
* Depicts the variation in the data distribution
* It is used to see how data is distributed. Output will be in the form of histogram 
 with the KDE (kernel density estimation ) line

```python
# A Distplot or distribution plot, depicts the variation in the data distribution
# displot - is used to see how data is distributed. Output will be in the form of histogram 
# with the KDE (kernel density estimation ) line

plt.rcParams["patch.force_edgecolor"] = True
sns.distplot(tips['total_bill'])
# Looking at the histogram we can say most of our bill is between 10 & 20 dollars
```

```python
# In case we dont want KDE to be there
sns.distplot(tips['total_bill'], kde = False)
```

```python
fig = plt.figure(figsize=(3,3), dpi=100)
sns.distplot(tips['total_bill'], kde = False)
```

```python
# label each bar in histogram
ax = sns.distplot(tips['total_bill'], kde = False)
for p in ax.patches:
 height = p.get_height() # get the height of each bar
 # adding text to each bar
 ax.text(x = p.get_x()+(p.get_width()/2), # x-coordinate position of data label, padded to be in the middle of the bar
 y = height+0.2, # y-coordinate position of data label, padded 0.2 above bar
 s = '{:.0f}'.format(height), # data label, formatted to ignore decimals
 ha = 'center') # sets horizontal alignment (ha) to center
```

# jointplot  
* It is used to get more than one plot together.  
* We can see the relationship between two variables in form of scatter or hexagonal plot

```python
#  On X axis we have Total Bill columns -  scatter plot & on Y axis we have Tips column
#  Here we can easily see the relationship between total bills that customers are paying and tips they are giving
sns.jointplot(x = 'total_bill',y = 'tip', data = tips)
# Here we can see higher the bill more the tips
```

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips, kind = 'hex')
# Very similar to scatter. It gets darker where we have concentration of points more or high desity
```

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips, kind = 'reg')
# To look at the linear fit or linear linse
```

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips, kind = 'kde')
# Very similar to scatter. It gets darker where we have concentration of points more or high desity
```

# Categorical Plots

## Bar Plots
1. BarPlot - a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.
2. Aggregate the categorical data based on some function. By default its by the mean. Visualsation by group by action.

```python
# SHows the mean value
# One is categorical Data & other is numerical data
sns.barplot(x = 'sex',y = 'total_bill', data = tips)
```

```python
# Standard Deviation shows variance  - whether more variance in bill of males or females
# estimator - by default is average
# We can have our functions also or any other inbuilt function
sns.barplot(x = 'sex',y = 'total_bill', data = tips, estimator=np.std)
```

```python
# countplot
# Count the number of occurances
sns.countplot('sex', data = tips)
```

# Strip Plot
1. In strip plot is scatter plot where one column is categorical and other continuous or numerical
2. By default jitter or noise is added. That is jitter = True
3. With jitter we can see the density little better

```python

sns.stripplot(x = 'day', y = 'total_bill',data = tips, jitter = False)
```

# Matrix Plots

2. Index name and column name should match up so that the cell value match up and relevant to both names
3. Index and columns both should be variables
4. example correlation data - tips.corr()

```python
# Correlation Data is used to plot heat maps
tips.corr()
```

```python
tips_corr = tips.corr()
```

```python
fig
sns.heatmap(tips_corr)
```

```python
# Annotation= True wil show th eactual values
# plt.subplots(figsize=(20,15))


ax = sns.heatmap(tips_corr, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top-0.5)
```
